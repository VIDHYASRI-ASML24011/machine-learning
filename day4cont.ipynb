```python
import joblib
import numpy as np

model = joblib.load("car_buy_model.pkl")

# User input
age = int(input("Enter Age: "))
income = int(input("Enter Income (Lakhs): "))
credit = int(input("Enter Credit Score: "))
exp = int(input("Enter Years of Experience: "))

new_data = np.array([[age, income, credit, exp]])
prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Prediction: Person WILL buy a car ")
else:
    print("Prediction: Person will NOT buy a car")
```
