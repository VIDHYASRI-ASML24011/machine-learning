```python
import joblib
import numpy as np

# Load model and polynomial transformer
model = joblib.load("poly_marks_model.pkl")
poly = joblib.load("poly_transform.pkl")
```

```python

# Take user input
hours = float(input("Enter number of hours studied: "))

# Convert to polynomial features

hours_array = np.array([[hours]])
hours_poly = poly.transform(hours_array)

# Predict marks
predicted_marks = model.predict(hours_poly)
print("Predicted Marks:", predicted_marks[0])
```
